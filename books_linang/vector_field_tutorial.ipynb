# How to plot vector field



```julia
using Plots
gr() #use GR module as a backend
xlims=[-2,2]
ylims=[-2,2]
```

## Define arrow function

```julia
function arrow!(ax, position, vector)
    vector = ([vector[1]], [vector[2]])
    quiver!(ax, position, quiver = vector, label = "label") #label does not work??
end
```

## Draw unit vectors from the origin

```julia
p = plot(xlims=[-2,2], ylims = [-2,2], aspect_ratio = 1)
ex = [1,0]
ey = [0,1]
r = ex + ey
arrow!(p, (0,0), ex)
arrow!(p, (0,0), ey)
arrow!(p, (0,0), r)
```

## Draw circle

```julia
ax = plot(aspect_ratio = 1) #initialization
for i in 0:0.2:2
    vector = [cos(i), sin(i)]
    arrow!(ax, (0,0),vector)
end
display(ax)
```
