# Newton-Raphson method 

## Aim
The method is used to find a solution to $f(x) = 0$ by using the derivative $f'(x)$.
Suppose that $\alpha$ is the true value of the solution and a adjacent value $x_k$. Then, the value $f(\alpha)$ can be approximated by: 

\begin{align}
f(\alpha) \approx f(x_k) + f'(x_k)(\alpha - x_k).
\end{align}

Here, more than second order terms are omitted.
Since $f(\alpha)=0$, 

\begin{align}
0 &\approx f(x_k) + f'(x_k)(\alpha - x_k) \\
\Rightarrow \alpha &\approx x_{k+1} = x_k - \frac{f(x_k)}{f'(x_k)}.
\end{align}
This is an approximated solution to $f(x) = 0$.
Assigning a initial value with $x_k$ and iterating, the approximated solution can be obtained.

```julia
using Plots
```

```julia
function calc_root_by_newton(a)
    n_steps = 10^3
    xk = a
    ϵ  = 10^-9
    errors = []
    for i in 1:n_steps
        xk_suc = xk - (xk^2 - a) / 2xk
        err = abs(xk_suc - xk)
        push!(errors, err) 
        println(xk_suc)
        if ϵ > err
            break
        end
        xk = xk_suc
    end
    return xk, errors
end
```

```julia
xk, errors = calc_root_by_newton(10000)
```
